# Производные

В этом задании вы научитесь использовать производную для решения несложных вычислительных задач.

```python
import numpy as np
from matplotlib import pyplot as plt
```

Для разогрева давайте найдём точки оптимума функции $f(x) = \frac{x^3}{x^4 + 1}$.

```python
# уменьшим копипаст и введём функцию
def f(x):
    return x ** 3 / (x ** 4 + 1)

x = np.arange(-5, 5, 0.05)
y = f(x)

plt.figure(figsize=(12, 6))
plt.title('$f(x) = x^3 / (x^4 + 1)$')
plt.plot(x, y, lw=2)
plt.grid(True)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```

Посчитаем производную:

$$f'(x) = \frac{\left(x^3\right)' \left(x^4 + 1\right) - x^3 \left(x^4 + 1\right)'}{\left(x^4 + 1\right)^2} = \frac{3 x^2 - x^6}{\left(x^4 + 1\right)^2}$$

И посмотрим на её график:

```python
def df(x):
    return (3 * x ** 2 - x ** 6) / (x ** 4 + 1) ** 2

y = df(x)

plt.figure(figsize=(12, 6))
plt.title("$f'(x) = (3 x^2 - x^6) / (x^4 + 1)^2$")
plt.plot(x, y, lw=2)
plt.grid(True)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```

Действительно, все точки оптимума являются нулями производной (обратное не совсем верно: посмотрите на точку $x = 0$).

Нули производной можно было бы найти и аналитически, но мы же здесь занимаемся численными методами, верно? Вспомним, как мы искали решения уравнений вида $f(x) = 0$.

```python
mask = (np.abs(x) > 1) & (np.abs(y) < 0.02)

plt.figure(figsize=(12, 6))
plt.title("$f'(x) = (3 x^2 - x^6) / (x^4 + 1)^2$")
plt.plot(x, y, lw=2)
plt.scatter(x[mask], y[mask])
plt.grid(True)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```

Напоследок давайте проработаем формулу касательной в точке:

$$g(x) = f(x_0) + (x - x_0) f'(x_0).$$

Более конкретно, нарисуем вместе график нашей функции $f$, касательной $g$ в точке $x_0 = 0.3$ и (примерные) точки их пересечения.

```python
def tangent(x, x0):
    return f(x0) + (x - x0) * df(x0)

x0 = 0.3
y = f(x)
g = tangent(x, x0)
mask = np.abs(y - g) < 0.009

plt.figure(figsize=(12, 6))
plt.title("$f(x) = x^3 / (x^4 + 1)$")
plt.plot(x, y, lw=2)
plt.plot(x, g, lw=2, color='green')
plt.scatter(x[mask], y[mask])
plt.grid(True)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```

Заметим, что "касательная" пересекает график в нескольких точках. Почему же она так называется?

Имеется в виду локальное поведение &mdash; если посмотреть на окрестность точки $x_0$, $g$ действительно просто касается $f$.

```python
x = np.arange(0.2, 0.4, 0.01)
y = f(x)
g = tangent(x, x0)

plt.figure(figsize=(8, 6))
plt.title('$f(x) = x^3 / (x^4 + 1)$')
plt.plot(x, y, lw=2, label='Функция')
plt.plot(x, g, lw=2, color='green', label='Касательная')
plt.grid(True)
plt.legend()
plt.show()
```
