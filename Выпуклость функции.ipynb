```python
import sympy
x,y,z = sympy.symbols('x y z')

```

```python
f = (x+3)**(4/3)*(1-x)**(2/3)
```

```python
sympy.plotting.plot(f, xlim=(-4,0), ylim=(-10,10))
```

```python
f2 = -((x+2)/(2*x+7))**3
sympy.plotting.plot(f2, xlim=(-8,-3), ylim=(-10,10))
```

```python
f2 = (2-x)/((x+1)**(2/3))
sympy.plotting.plot(f2, xlim=(-50,50), ylim=(-50,50))
```

```python
sympy.solve(f2.diff().diff())
```
